```python
# 將部份所需要的 Column 截出來，建 Derived Table
short = '''select coid
,fnamec
,inamec
,mkt
,tejinm2_c
,tejinm3_c
,tejinm4_c
,tejico2
,tejico3
,tejico4
from TEJ_STOCK_DATA
where mkt in ('TSE','OTC')
;'''
df = pd.read_sql_query(short, con=engine)
```

```python
df.rename(columns={'coid':'ID'
                    ,'fnamec':'F_NAME'            
                    ,'inamec':'NAME'
                    ,'mkt':'MKT'
                    ,'tejinm2_c':'INDUSTRY_NAME_1'
                    ,'tejinm3_c':'INDUSTRY_NAME_2'
                    ,'tejinm4_c':'TSE_INDUSTRY_NAME'
                    ,'tejico2':'INDUSTRY_ID_1'      
                    ,'tejico3':'INDUSTRY_ID_2'
                    ,'tejico4':'TSE_INDUSTRY_ID'}, 
         inplace=True)
```
